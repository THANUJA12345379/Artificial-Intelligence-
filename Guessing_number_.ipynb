```python
def guess_number():
    print("Think of a number between 1 and 100, and I will try to guess it!")
    input("Press Enter when you're ready...")

    low = 1
    high = 100

    while low <= high:
        guess = (low + high) // 2
        print(f"Is your number {guess}?")
        response = input("Enter 'higher', 'lower', or 'correct': ").lower()

        if response == 'h':
            low = guess + 1
        elif response == 'l':
            high = guess - 1
        elif response == 'c':
            print("Yay! I guessed your number!")
            break
        else:
            print("Please enter a valid response: 'higher', 'lower', or 'correct'.")

guess_number()

```
